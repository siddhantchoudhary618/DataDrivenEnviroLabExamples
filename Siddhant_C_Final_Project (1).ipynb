```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('gym_members_exercise_tracking.csv')

data
```

```python
#Questions

print("1. What is the average calories burned across all workout sessions?")
print("2. Which workout type results in the highest average calories burned?")
print("3. How does session duration correlate with calories burned?")
print("4. What is the average BMI grouped by experience level?")
print("5. Which gender has a higher average workout frequency?")
```

```python
# Question 1: Average calories burned
average_calories = data['Calories_Burned'].mean()
average_calories
```

```python
# Question 2: Highest average calories burned by workout type
highest_avg_calories = data.groupby('Workout_Type')['Calories_Burned'].mean().idxmax()
highest_avg_calories
```

```python
# Question 3: Correlation between session duration and calories burned
correlation_duration_calories = data['Session_Duration (hours)'].corr(data['Calories_Burned'])
correlation_duration_calories
```

```python
# Question 4: Average BMI by experience level
average_bmi_experience = data.groupby('Experience_Level')['BMI'].mean()

# display
average_bmi_experience_data = average_bmi_experience.reset_index()
average_bmi_experience_data.rename(columns={'BMI': 'Average BMI'}, inplace=True)
average_bmi_experience_data
```

```python
# Question 5: Average Workout Frequency by Gender
average_workout_frequency_gender = data.groupby('Gender')['Workout_Frequency (days/week)'].mean()

#display
average_workout_frequency_gender_data = average_workout_frequency_gender.reset_index()
average_workout_frequency_gender_data.rename(columns={'Workout_Frequency (days/week)': 'Average Workout Frequency'}, inplace=True)
average_workout_frequency_gender_data
```

```python
# Sort by BMI in ascending order
sorted_bmi_asc = data.sort_values(by='BMI', ascending=True)
sorted_bmi_asc
```

```python
# Sort by Calories_Burned in descending order
sorted_calories_desc = data.sort_values(by='Calories_Burned', ascending=False)
sorted_calories_desc
```

```python
# Count of Workout Types
workout_type_counts_data = data.groupby('Workout_Type').size().reset_index(name='Count')
workout_type_counts_data.rename(columns={'Workout_Type': 'Workout Type'}, inplace=True)
workout_type_counts_data
```

```python
# Maximum Water Intake by Gender
max_water_intake_gender_data = data.groupby('Gender', as_index=False)['Water_Intake (liters)'].max()
max_water_intake_gender_data.rename(columns={'Water_Intake (liters)': 'Max Water Intake (liters)'}, inplace=True)
max_water_intake_gender_data
```

```python
# Average Session Duration by Experience Level
avg_session_duration_experience_data = data.groupby('Experience_Level', as_index=False)['Session_Duration (hours)'].mean()
avg_session_duration_experience_data.rename(columns={'Session_Duration (hours)': 'Average Session Duration (hours)'}, inplace=True)
avg_session_duration_experience_data
```

```python
# Average workout frequency by gender
average_workout_frequency_gender.plot(kind='bar', color=['blue', 'pink'], title='Workout Frequency by Gender')
plt.xlabel('Gender')
plt.ylabel('Workout Frequency (days/week)')
plt.show()
```

```python
# Average BMI by experience level
average_bmi_experience.plot(kind='bar', color='green', title='Average BMI by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average BMI')
plt.show()
```

```python
# Scatter plot: Session duration vs. Calories burned
plt.scatter(data['Session_Duration (hours)'], data['Calories_Burned'], color='red')
plt.title('Session Duration vs. Calories Burned')
plt.xlabel('Session Duration (hours)')
plt.ylabel('Calories Burned')
plt.show()
```

```python
# Scatter plot: BMI vs. Fat Percentage
plt.scatter(data['BMI'], data['Fat_Percentage'], color='blue')
plt.title('BMI vs. Fat Percentage')
plt.xlabel('BMI')
plt.ylabel('Fat Percentage')
plt.show()
```

```python
# Quantitative variable: Calories_Burned
quantitative_variable = data['Calories_Burned']
```

```python
import numpy as np

def random_sample_average(data, sample_size):
    """Function to take a random sample of a specified size from the data and return its average."""
    sample = np.random.choice(data, size=sample_size, replace=True)
    return np.mean(sample)

# Test the function with a sample size of 30
random_sample_average(quantitative_variable, 30)
```

```python
def repeated_sampling(data, sample_size, iterations=10000):
    """Function to sample the data 10,000 times, each with the specified sample size. Returns an array of average values."""
    averages = []
    for i in range(iterations):
        avg = random_sample_average(data, sample_size)
        averages.append(avg)
    return np.array(averages)

# Test the function with a sample size of 30
sampled_averages = repeated_sampling(quantitative_variable, sample_size=30)
len(sampled_averages)
```

```python
# Generate 10,000 averages for sample size of 200
sample_size = 200
sampled_averages_200 = repeated_sampling(quantitative_variable, sample_size=sample_size)
```

```python
#Table for the sampled averages
averages_data = pd.DataFrame(sampled_averages_200, columns=['Sampled_Average'])
averages_data.head()
```

```python
# Plot the distribution of sampled averages
plt.hist(averages_data['Sampled_Average'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Sampled Averages (Sample Size = 200)')
plt.xlabel('Sampled Average of Calories Burned')
plt.ylabel('Frequency')
plt.show()
```
